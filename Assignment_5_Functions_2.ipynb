## Task1:
You need to write a function in python that return all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included).
- Return the obtained numbers in a comma-separated sequence on a single line. 

```python
def print_seven_divisible():
    for i in range(2000, 3200):
        if (i % 7 == 0) and (i % 5 != 0):
            print(f'{i}', end=',')
```

```python
print_seven_divisible()
```

## Task2:
You need code a function that calculates and returns the value according to the given formula:
- P = Square root of [(2 * A * B)/C]
Following are the fixed values of A and B:
- A is 50. B is 30.
The values of the literal C should be taken as console input to your program in a comma-separated sequence.
- Example
    - Let us assume the following comma separated input sequence is given to the program: 100,150,180
    - The output of the program should be: 18,22,24

```python
import math

def calculate(c, a=50, b=30):
    if c == 0:
        return "Division by zero is not allowed."
    result = math.sqrt((2 * a * b) / c)
    return round(result)
```

```python
numbers = input('Enter three number for c seperated by qoma: ')
number_list = numbers.split(',')
for number in number_list:
    print(f'{str(calculate(int(number)))}')
```

## Task3:
You need to write a function that takes a comma separated sequence of words as input and prints the words in a comma-separated sequence after sorting them alphabetically.
- Suppose the following input is supplied to the program: without,hello,bag,world
- Then, the output should be: bag,hello,without,world 


```python
def sort_words(input_string):
    words = input_string.split(',')
    
    words.sort()
    
    return ','.join(words)
```

```python
input_string = input("Enter a comma-separated sequence of words: ")
sorted_words = sort_words(input_string)
print("Sorted words:", sorted_words)
```

## Task 3
You need to write a program that takes sequence of lines as input and prints the lines after making all characters in the sentence capitalized.
- Suppose the following input is supplied to the program:
    - Hello world
    - Practice makes perfect
- Then, the output should be:
    - HELLO WORLD
    - PRACTICE MAKES PERFECT 

```python
def capitalize_lines():
    # Take multiple lines of input from the user
    while True:
        try:
            line = input()
            if line == '':
                break
            # Print the line in uppercase
            print(line.upper())
        except EOFError:
            break
```

```python
print("Enter multiple lines. Press Enter on an empty line to stop:")
capitalize_lines()
```

## Task5:
You need to write a function that counts the number of vowels in a given sentence as input from console.
- Suppose the following input is supplied to the program:
    - Practice makes perfect
Then, the output should be:
  - a appeared 2 times
  - e appeared 5 times
  - i appeared 1 times
  - o appeared 2 times
  - u appeared 0 time
  - PRACTICE MAKES PERFECT

```python
def count_vowels(data):
    vowel_list = ['a','e','i','o','u']
    vowel_counts = [0,0,0,0,0]
    
    for char in data:
        if char.lower() in vowel_list:
            index = vowel_list.index(char.lower())
            vowel_counts[index]+=1
    return vowel_list, vowel_counts
```

```python
count_vowels('Hello world Practice makes perfect')
```

```python
def count_vowels_in_lines():
    vowel_list = ['a','e','i','o','u']
    vowel_counts = [0,0,0,0,0]
    
    while True:
        try:
            line = input()
            if line == '':
                return vowel_list, vowel_counts
            for char in line:
                if char.lower() in vowel_list:
                    index = vowel_list.index(char.lower())
                    vowel_counts[index]+=1
        except EOFError:
            break
```

```python
vowels, vowels_count = count_vowels_in_lines()
for index, vowel in enumerate(vowels):
    print(f'{vowel} appeared {vowels_count[index]} times')
```

## Task6:
You need write a function that traces and makes a list of all such numbers from1000 to 3000 in which all the digits are even numbers.

```python
def even_numbers():
    even_numbers = []
    for number in range(1000,3000+1):
        if number % 2 == 0:
            even_numbers.append(number)
    return even_numbers
```

## Task7:
You need to write a code which accepts a sequence of comma separated 4 digit binary numbers as its input and then check whether they are divisible by 5 or not.
The numbers that are divisible by 5 are to be printed in a comma separated
sequence.
Example:
0100,0011,1010,1001
Then the output should be:
1010
Reference: https://www.datacamp.com/.../python-data-type-conversion

```python
def check_divisible_by_5(binary_numbers):
    divisible_by_5 = []
    for binary in binary_numbers:
        decimal = int(binary, 2)
        if decimal % 5 == 0:
            divisible_by_5.append(binary)
    return ','.join(divisible_by_5)
```

```python
input_numbers = input("Enter comma-separated 4-digit binary numbers: ")
binary_numbers = input_numbers.split(',')
divisible_by_5 = check_divisible_by_5(binary_numbers)
print(divisible_by_5)
```

## Task8:
Write a program that accepts a sentence and calculate the number of letters and digits.
- Suppose the following input is supplied to the program:
    - hello world! 123
- Then, the output should be:
    - LETTERS 10
    - DIGITS 3 

```python
def calculate_letter_count(data):
    letter_count = 0
    digit_count = 0
    for char in data:
        if char.isalpha():
            letter_count+=1
        elif char.isdigit():
            digit_count+=1
    return letter_count, digit_count
```

```python
sentence = input('Enter sentence to count letter and digit: ')
letter_count, digit_count = calculate_letter_count(sentence)
print(f'Letters {letter_count}')
print(f'Digit {digit_count}')
```

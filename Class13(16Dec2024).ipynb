```python
class Student():
    # initializer / constrcutor
    def __init__(self,name, age, city, course='Python'):
        self.name = name
        self.age = age
        self.city = city
        self.course  = course
        
    
    # functions / Methods
    def apper_in_exam(self):
        print(f"{self.name} is appearing in exam")
    
    # functions / Methods
    def pay_fees(self):
        print(f"{self.name} is paying fee")
             
```

```python
s4 = Student()
```

```python
s4 = Student('Taha', 20, "Karachi", 'Python')
```

```python
s5 = Student('Wali', 23,'Lahore')
```

```python
s4.apper_in_exam()
```

```python
s5.apper_in_exam()
```

```python
class Car():
    def __init__(s,model, make, color='Black'):
        s.model =model
        s.make = make
        s.color = color
        s.ac = 'Dawlance'
    def describe_car(s):
        print(f"""
             Model Name: {s.model}
             Car Make:   {s.make}
             Car Color:  {s.color}
             Car Ac: {s.ac}
             
             """)
```

```python
c1 = Car('2024', "Honda", 'Black')
c2 = Car('2025', "Honda", 'White')
```

```python
c1.describe_car()
```

```python
c2.describe_car()
```

```python
Car.describe_car(c2)
```

```python
Car.describe_car(c1)
```
